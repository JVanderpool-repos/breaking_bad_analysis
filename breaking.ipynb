```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
breaking_bad_df = pd.read_csv('breaking_bad.csv', encoding='latin1')

# Combine the 'Season' and 'Episode' columns into one column
breaking_bad_df['Season_Episode'] = breaking_bad_df['Season'].astype(str) + '/' + breaking_bad_df['Episode'].astype(str)

# Handle missing values in the 'U.S. viewers_million' column by filling with the median
breaking_bad_df['U.S. viewers_million'].fillna(breaking_bad_df['U.S. viewers_million'].median(), inplace=True)

# Convert the 'Date' column to datetime for better plotting
breaking_bad_df['Date'] = pd.to_datetime(breaking_bad_df['Date'], format='%d-%b-%y')

# Extract season from the 'Season_Episode' column
breaking_bad_df['Extracted_Season'] = breaking_bad_df['Season_Episode'].str.extract(r'(\d+)/')[0].astype(int)

# Plot IMDb ratings over time with different colors for each extracted season
plt.figure(figsize=(12, 6))
for season in breaking_bad_df['Extracted_Season'].unique():
    season_data = breaking_bad_df[breaking_bad_df['Extracted_Season'] == season]
    plt.plot(season_data['Date'], season_data['Rating_IMDB'], marker='o', linestyle='-', label=f'Season {season}')
plt.title('IMDb Ratings Over Time', fontsize=16)
plt.xlabel('Air Date', fontsize=12)
plt.ylabel('IMDb Rating', fontsize=12)
plt.legend(title='Season')
plt.grid(True)
plt.show()

# Plot U.S. viewership by season
plt.figure(figsize=(12, 6))
seasons = breaking_bad_df['Extracted_Season'].unique()
viewership = [breaking_bad_df[breaking_bad_df['Extracted_Season'] == season]['U.S. viewers_million'] for season in seasons]
plt.boxplot(viewership, labels=seasons, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('U.S. Viewership by Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Viewership (Millions)', fontsize=12)
plt.grid(True, axis='y')
plt.show()

# Plot distribution of episode durations
plt.figure(figsize=(12, 6))
plt.hist(breaking_bad_df['Duration_mins'], bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Episode Durations', fontsize=16)
plt.xlabel('Duration (Minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```
